```python
from bs4 import BeautifulSoup as bs
import urllib2
#import string
import pandas as pd
```

```python
# This function reads the url and the option- 'bat' or 'bowl' and creates a dataframe with corresponding data

def f_get_stats(url,stt): # pass the url and the option stt='bat'/'bowl'
    page = urllib2.urlopen(url) 
    soup = bs(page.read()) # Extract entire page
    tbls=soup.find_all('table') # Extract all tables
    if stt=='bat': 
        bat=tbls[0].find_all('tr') # extract the first table ( for Batting )
    elif stt=='bowl':
        bat=tbls[1].find_all('tr') # extract the first table ( for Bowling )
    lst=[[] for i in range((len(bat)))]

    for k,i in enumerate(bat):
        if k==0:
            head=i.find_all('th') # the first line has the heading
            headings=[str(ff.text) for ff in head]
        else :
            lst[k]=[str(j.text) for j in i.find_all('td')]

    #lst[0] is empty
    rws=[lst[i][0] for i in range(1,len(lst))] # The first column of list stores the row name
    clmns=list(headings)[1::] # the column names
    new_lst=[i[1::] for i in lst[1::]]
    return pd.DataFrame(new_lst,index=rws,columns=clmns)
```

```python
url='http://www.espncricinfo.com/ci/content/player/45789.html'
d1=f_get_stats(url,'bat')
d2=f_get_stats(url,'bowl')
url='http://www.espncricinfo.com/ci/content/player/35320.html'
t1=f_get_stats(url,'bat')
t2=f_get_stats(url,'bowl')
```
